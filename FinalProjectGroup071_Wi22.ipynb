## Background & Prior Work

The stock market is comprised of and measured in stock market indexes [[1]](https://www.forbes.com/advisor/investing/stock-market-index/). These indexes allow investors the opportunity to track their investment’s health, and whether a stock is on the rise or downfall. These indexes can be composed of several stock types such as agriculture, technology, real estate, and more. The two popular stock indexes are the Dow Jones Industrial Average (DJIA), and Standard & Poor's 500 Index (S&P 500) [[2]](https://www.investopedia.com/ask/answers/difference-between-dow-jones-industrial-average-and-sp-500/#:~:text=The%20DJIA%20tracks%20the%20stock,the%20stock%20markets%20in%20general.). Each index takes measurements of popular company stocks and reports their averages. 
<br><br>
For our research, we will be using a stock index known as QQQ to measure technology-related stock prices. The QQQ is a combination of multiple notable technology companies and acts as an overview for general trends in technology. To analyze stock in physical entertainment, we will be observing the Las Vegas Sands Corporation (LVS) stock prices. Las Vegas Sands Corporation is a casino and resort company based in Paradise, Nevada. Finally, we will be comparing both stock datasets to the WHO COVID-19 Global Data Report. This data report provided by the World Health Organization provides a global count of new and cumulative COVID-19 cases. We will only look at COVID-19 data in the United States.
<br><br>
COVID-19 is a disease discovered in December 2019, in Wuhan, China [[3]](https://www.cdc.gov/coronavirus/2019-ncov/your-health/about-covid-19/basics-covid-19.html). This highly contagious virus spreads through droplets via the mouth and other manners. Symptoms include fever, cough, headache, body aches, and more [[4]](https://www.cdc.gov/coronavirus/2019-ncov/symptoms-testing/symptoms.html). It is recommended to prevent the spread of the virus to wear medical-grade masks when possible that are well-fitted [[5]](https://www.cdc.gov/coronavirus/2019-ncov/prevent-getting-sick/masks.html). Frequent hand washing and general sanitization of shared and high-touch surfaces are also recommended. Beyond these everyday precautions, the CDC and other government agencies recommend receiving the COVID-19 vaccine (a two-dose system) which lessens the likelihood of death and helps patients manage their symptoms. 
<br><br>
On the whole, to understand how COVID-19 caused a stock market crash in March 2020 is to understand what led up to this event. As more cases of COVID-19 were reported in the U.S., and several states and businesses were shutting down in early March, stock investors feared the unknown that was COVID-19 [[6]](https://www.thebalance.com/fundamentals-of-the-2020-market-crash-4799950). States such as California began declaring a state of emergency on March 4th, 2020 [[7]](https://calmatters.org/health/coronavirus/2021/03/timeline-california-pandemic-year-key-points/), followed by New York on March 7, 2020 [[8]]( https://www.investopedia.com/historical-timeline-of-covid-19-in-new-york-city-5071986). Finally, former President Trump declared a national state of emergency on March 13, 2020. Before March 13, the Dow Jones Industrial Average (DJIA) had already seen a significant point drop. However, three days after the national emergency declaration, March 16, 2020, saw a record low for the DJIA at a 12.93% point fall [[6]](https://www.thebalance.com/fundamentals-of-the-2020-market-crash-4799950). After this, a recession was triggered for the rest of 2020. 
<br><br>
According to a prior study, stocks such as technology, healthcare, and others saw the most positive equity value [[9]](https://www.ncbi.nlm.nih.gov/pmc/articles/PMC7343658/). On the other end, stocks such as entertainment and real estate saw low returns [[9]](https://www.ncbi.nlm.nih.gov/pmc/articles/PMC7343658/). That said, our group would like to observe the stocks of U.S.-based companies to determine if COVID-19 did influence the particular stock markets of technology and entertainment.

```python
# generate QQQ data in the form of date, change since previous day
QQQ_data = []
for i in range(1,len(QQQ_df_2020)):
    QQQ_data.append([QQQ_df_2020['Date'].iloc[i],QQQ_df_2020['Close'].iloc[i-1]-QQQ_df_2020['Close'].iloc[i]])
QQQ_data = pd.DataFrame(QQQ_data, columns = ['Date', 'Closing Change'])
```

```python
ax = sns.lineplot(x='Date', y='Closing Change', data=QQQ_data)
x_ticks = ax.set_xticks([i*30 for i in range(10)])
ax.set(title="QQQ Closing Change Performance In 2020")
```

```python
# generate data in the form of date, change since previous day
LVS_data = []
for i in range(1,len(LVS_df_2020)):
    LVS_data.append([LVS_df_2020['Date'].iloc[i],LVS_df_2020['Close'].iloc[i-1]-LVS_df_2020['Close'].iloc[i]])
LVS_data = pd.DataFrame(LVS_data, columns = ['Date', 'Closing Change'])
```

```python
ax = sns.lineplot(x='Date', y='Closing Change', data=LVS_data, color='red')
x_ticks = ax.set_xticks([i*30 for i in range(10)])
ax.set(title="LVS Closing Change Performance In 2020")
```

```python
fig, ax = plt.subplots()
sns.lineplot(x='Date', y='Closing Change', data=QQQ_data)
plt.ylabel('Closing Change QQQ', size=15)
ax2 = ax.twinx()
sns.lineplot(x='Date', y='Closing Change', data=LVS_data, color='r')
plt.ylabel('Closing Change LVS', size=15)
plt.xlabel('Date', size=15)
x_ticks = ax.set_xticks([i*30 for i in range(10)])
```

```python
# Compare number of positive deltas to negative deltas for QQQ dataset
QQQ_pos_deltas = sum(QQQ_data['Closing Change'] > 0)
QQQ_neg_deltas = sum(QQQ_data['Closing Change'] < 0)
QQQ_pos_neg_ratio = QQQ_pos_deltas / QQQ_neg_deltas
print('Ratio of QQQ positive deltas to negative deltas for QQQ dataset:', QQQ_pos_neg_ratio)
```

```python
# Compare number of positive deltas to negative deltas for LVS dataset
LVS_pos_deltas = sum(LVS_data['Closing Change'] > 0)
LVS_neg_deltas = sum(LVS_data['Closing Change'] < 0)
LVS_pos_neg_ratio = LVS_pos_deltas / LVS_neg_deltas
print('Ratio of positive deltas to negative deltas for LVS dataset:', LVS_pos_neg_ratio)
```

```python
# Compare means of positive delta means to negative delta means for QQQ dataset
QQQ_pos_mean = abs(QQQ_data[QQQ_data['Closing Change'] > 0]['Closing Change'].mean())
QQQ_neg_mean = abs(QQQ_data[QQQ_data['Closing Change'] < 0]['Closing Change'].mean())
QQQ_pos_neg_mean_ratio = QQQ_pos_mean / QQQ_neg_mean
print('Ratio of QQQ positive changes in value to negative changes', QQQ_pos_neg_mean_ratio)
```

```python
# Compare means of positive delta means to negative delta means for LVS dataset
LVS_pos_mean = abs(LVS_data[LVS_data['Closing Change'] > 0]['Closing Change'].mean())
LVS_neg_mean = abs(LVS_data[LVS_data['Closing Change'] < 0]['Closing Change'].mean())
LVS_pos_neg_mean_ratio = LVS_pos_mean / LVS_neg_mean
print('Ratio of LVS positive changes in value to negative changes', LVS_pos_neg_mean_ratio)
```

In order to analyze the revised stock data, what we decided to do is use the change in values from the closing price of the current day to the day before. In this manner, what we would end up with is the “strength” of the stock by seeing how it changed from day to day. By taking the ratio of positive to negative days, as well as the mean of the positive days and mean of negative days of each respective stock (QQQ and LVS), what we hope to do is determine which of the two was stronger in the sense of having more positive change days, or having larger “positive” days or lesser “negative” days to denote stock strength. From our findings, what we found was that the QQQ had less positive days compared to negative days in comparison to LVS, however the ratio of positive changes in value to negative changes was much bigger compared to that of LVS’s ratio of positive changes in value to negative changes. This is partially in line with our hypothesis as it demonstrates how the QQQ for the most part increased at a larger rate from day to day in comparison to LVS, despite the fact that it has a lesser ratio of positive days to negative days. Particular confounding variables which could particularly interfere with these results is the stretch of data we are using to analyze. The data covers the time period of before the COVID crash which could potentially account for the unexpectedness of QQQ having less positive days than negative.
